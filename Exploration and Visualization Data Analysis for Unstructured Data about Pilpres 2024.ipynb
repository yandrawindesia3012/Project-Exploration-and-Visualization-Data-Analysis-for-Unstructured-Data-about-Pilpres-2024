# Scrapping Data from Website

```python
import requests as req
from bs4 import BeautifulSoup as bs
from datetime import datetime
import csv
import pandas as pd

hades = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}

def scrape_detik(hal):
  global hades
  a = 1
  data = []  # List to store the scraped data
  for page in range(1, hal):
    url = f'https://www.detik.com/search/searchall?query=pilpres&sortby=time&fromdatex=01/03/2023&todatex=31/05/2023&siteid={page}'
    ge = req.get(url, headers=hades).text
    sop = bs(ge, 'lxml')
    li = sop.find('div', class_='list media_rows list-berita')
    lin = li.find_all('article')
    for x in lin:
      link = x.find('a')['href']
      date = x.find('a').find('span', class_='date').text.replace('WIB','').replace('detikNews','').split(',')[1]
      headline = x.find('a').find('h2').text
      ge_ = req.get(link, headers=hades).text
      sop_ = bs(ge_,'lxml')
      content = sop_.find_all('div',class_='detail__body-text itp_bodycontent')
      for x in content:
        x = x.find_all('p')
        y = [y.text for y in x]
        content_ = ''.join(y).replace('\n','').replace('ADVERTISEMENT','').replace('SCROLL TO RESUME CONTENT','')
        print(f'done[{a}] > {headline[0:10]}')
        a += 1
        data.append([headline, date, link, content_])

  # Save the data to a CSV file
  df = pd.DataFrame(data, columns=['Headline', 'Date', 'Link', 'Content'])
  df.to_csv('preprocessed_data.csv', index=False)

scrape_detik(1000)
```

```python
import requests
from bs4 import BeautifulSoup

# List of URLs
urls = ['https://www.detik.com/search/searchall?query=pilpres&sortby=time&fromdatex=01/03/2023&todatex=31/05/2023&siteid={page}']

# Empty list to store extracted text
extracted_text = []

# Iterate over the URLs
for url in urls:
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36'}
        page = requests.get(url, headers=headers)
        soup = BeautifulSoup(page.content, 'html.parser')
        content = soup.find(attrs={'class': 'td-post-content'})

        # Extract text and append to the list
        extracted_text.append(content.get_text() if content else "")
    except Exception as e:
        print(f"Error processing URL: {url}")
        print(e)

# Print the extracted text
for text in extracted_text:
    print(text)
    print('---')

df.info()
```

```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Load the data from CSV
df = pd.read_csv('tokenized_data.csv')

# Download stopwords if not already downloaded
import nltk
nltk.download('stopwords')

# Get the list of stopwords
stopwords_list = stopwords.words('english')

# Function to remove stopwords from a text
def remove_stopwords(text):
    tokens = word_tokenize(text)
    filtered_tokens = [token for token in tokens if token.lower() not in stopwords_list]
    return ' '.join(filtered_tokens)

# Apply stopwords removal to the 'Content' column
df['Content_No_Stopwords'] = df['Content'].apply(remove_stopwords)

# Print the content without stopwords
for index, row in df.iterrows():
    print(f"Content without Stopwords for Row {index+1}:")
    print(row['Content_No_Stopwords'])
    print()

# Save the data without stopwords to a new CSV file
df.to_csv('data_without_stopwords.csv', index=False)
```

```python
import pandas as pd
from collections import Counter
import re

# Load the data from CSV
df = pd.read_csv('data_without_stopwords.csv')

# Combine all the content into a single string
all_content = ' '.join(df['Content'])

# Function to calculate the probability of a word
def calculate_probability(word, tokens_count):
    return tokens_count[word] / sum(tokens_count.values())

# Function to generate possible corrections for a word
def generate_corrections(word):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    deletes = [left + right[1:] for left, right in splits if right]
    inserts = [left + letter + right for left, right in splits for letter in letters]
    replaces = [left + letter + right[1:] for left, right in splits if right for letter in letters]
    transposes = [left + right[1] + right[0] + right[2:] for left, right in splits if len(right) > 1]
    return set(deletes + inserts + replaces + transposes)

# Function to get the most probable correction for a word
def get_correction(word, tokens_count):
    possible_corrections = generate_corrections(word)
    return max(possible_corrections, key=lambda w: calculate_probability(w, tokens_count))

# Function to correct the content
def correct_content(content):
    tokens = re.findall(r'\b\w+\b', content.lower())
    tokens_count = Counter(tokens)
    corrected_tokens = []
    for token in tokens:
        if token not in tokens_count:
            corrected_tokens.append(get_correction(token, tokens_count))
        else:
            corrected_tokens.append(token)
    return ' '.join(corrected_tokens)

# Apply spell correction to the 'Content' column
df['Corrected_Content'] = df['Content'].apply(correct_content)

# Print the corrected content
for index, row in df.iterrows():
    print(f"Corrected Content for Row {index+1}:")
    print(row['Corrected_Content'])
    print()

# Save the corrected data to a new CSV file
df.to_csv('corrected_data.csv', index=False)
```

```python
import pandas as pd
import base64

# Load the data from CSV
df = pd.read_csv('/content/spell_checked_data.csv')

# Encode the content using base64
df['Content_Encoded'] = df['Corrected_Content'].apply(lambda x: base64.b64encode(x.encode('utf-8')).decode('utf-8'))

# Print the encoded content
for index, row in df.iterrows():
    print(f"Encoded Content for Row {index+1}:")
    print(row['Content_Encoded'])
    print()

# Decode the content using base64
df['Content_Decoded'] = df['Content_Encoded'].apply(lambda x: base64.b64decode(x).decode('utf-8'))

# Print the decoded content
for index, row in df.iterrows():
    print(f"Decoded Content for Row {index+1}:")
    print(row['Content_Decoded'])
    print()

# Save the encoded and decoded data to a new CSV file
df.to_csv('encoded_decoded_data.csv', index=False)
```

# Menghapus Singkatan

```python
import pandas as pd

# Load the data from CSV
df = pd.read_csv('/content/corrected_data.csv')

# Load the slang dictionary from CSV
slang_df = pd.read_csv('slang_dictionary.csv')

# Create a list of slang words
slang_words = slang_df['Slang'].tolist()

# Define a function to remove slang words
def remove_slang(text):
    words = text.split()
    cleaned_words = [word for word in words if word.lower() not in slang_words]
    cleaned_text = ' '.join(cleaned_words)
    return cleaned_text

# Apply the remove_slang function to the 'Content_Spell_Checked' column
df['Content_Cleaned'] = df['Corrected_Content'].apply(remove_slang)

# Print the cleaned content
for index, row in df.iterrows():
    print(f"Cleaned Content for Row {index+1}:")
    print(row['Content_Cleaned'])
    print()

# Save the cleaned data to a new CSV file
df.to_csv('cleaned_data.csv', index=False)
```
